## **Step 6: Cleaning the departure report dataset**

#### Required Data clean up
* Find and remove duplicate rows
* Replace missing values in the **carrier** column with the corresponding values found above.
* Change data types of **carrier** to **string** and **Data_dte** to **date**
* Convert column names to lower case and rename a few columns for readability.

### *Departure Report Dataset*

#### Looking at duplicate values

```python
# Looking at duplicate rows in the Departures report 

df1[df1.duplicated()]
```

There are no duplicate rows present in this dataset.

#### Converting the column names to lower case

```python
df1.columns = df1.columns.str.lower()
df1.head()
```

### Rename columns

```python
df1.rename(columns={'data_dte':'date','airlineid':'airline_id','carriergroup':'carrier_group'}, inplace = True)
df1.head()
```

### Changing data types

```python
# Changing date from object data type to a date format

df1["date"] = pd.to_datetime(df1["date"])
df1.head()
```

```python
# Changing carrier column from object to string

df1['carrier'] = df1['carrier'].astype(str)
```

### Replacing null values

#### Departure Report Dataset

```python
# Replacing null values in the carrier column of the dataset with values correseponding to the airline id.

# Filling missing values for 20414 with "OW" - see previous cell

df1.loc[df1["airline_id"] == 20414, ["carrier"]].fillna("OW", inplace=True)
```

```python
# Filling missing values for 20415 with "XG" - see previous cell

df1.loc[df1["airline_id"] == 20415, ["carrier"]].fillna("XG", inplace=True)

# Another way- df1.loc[(df1["airline_id"] == 20415) & df1["carrier"].isnull(), ["carrier"]] = "XG"
```

```python
# Rechecking if the null values are replaced

df1.isnull().sum()
```

#### Replacing the values of 1 and 0 in the carrier_group with "Domestic"and "Foreign"

```python
# First let's change carrier_group column to string

df1['carrier_group'] = df1['carrier_group'].astype(str) 
```

```python
# Let's now change values in the column from 1 to Domestic.

df1['carrier_group'] = df1['carrier_group'].replace(str(1),'Domestic')
```

```python
# Let's change the values in the column from 0 to Foreign

df1['carrier_group'] = df1['carrier_group'].replace(str(0),'Foreign')
```

```python
# Let's look at the carrier_group column to verify the changed values

df1['carrier_group']
```
